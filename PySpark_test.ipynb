```
# Creating DataFrame:
data = [("James","","Smith","36636","M",60000),
        ("Michel","Rose","","40288","M",70000),
        ("Robert","","Williamss","42114","",40000),
        ("Maria","Anne","Jonnes","39192","F",50000),
        ("Jen","Mary","Brown","","F",0)]

columns = ["first name", "middle_name", "last_name","dob", "gender", "salary"]
df = spark.createDataFrame (data = data, schema = columns)

display(df)
```

```
# add new column
display(df.withColumn("bonus",df.salary*0.2)) 
```

```
# how to filter and sorting 
display(df.filter(df.dob == 36636))
```

```
display(df.filter(df.salary > 10000))
```

```
display(df.sort(df.salary ))
```

```
display(df.sort(df.salary.asc()))
```

```
display(df.sort(df.salary.desc()))
```

```
# group by and aggregate 

# aggregate : sum,avg,min,max
display(df.select("first name", "gender", "salary"))
```

```
# sum of salary based on geneder

display(df.groupBy("gender").sum("salary"))
```

```
# 
display(df.groupBy("gender").count())
```

```
display(df.groupBy("gender").min("salary"))
```

```
display(df.groupBy("gender").avg("salary"))
```

```
# joining and union

df1.join(df2, df1.df1_c1 =df2.df2_c1, "inner").show()
```

```
# Union: row by row-- schema should be same.
# in pysspark union will not drop duplicate but in SQL it drops all the duplicates.


df1.union(df2).show()
```
